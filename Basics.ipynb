```python
def age_converter(Age):
    if pd.isnull(Age) == True:
        return None
    else:
        first_num = Age.split(' ')[0]
        first_num = first_num.replace('+', '')
        return int(first_num)
            
        
        
```

```python
#here we add a column with the age converted into ints
data['int_age'] = data['Age'].apply(age_converter)
```

```python
data['int_age'].describe()
```

we have to be aware of the fact that the number of participants in this survey is less than 1000 it might mean our statistics won't be very robust. 
also, the maximum age in the survey is 60 and the median is 45 so it really covers  apsecfic part of the population that might not be representative of the population

```python
def salary_converter(salary):
    if pd.isnull(salary) == True:
        return None
    else:
        first_bracket = salary.split()[0]
        if first_bracket == 'Prefer':
            return None
        else:
            first_bracket = first_bracket.replace('$', '')
            first_bracket = first_bracket.replace(',','')
            return int(first_bracket)
        
```

```python
data['int_income']= data['How much total combined money did all members of your HOUSEHOLD earn last year?'].apply(salary_converter)

            
```

```python
data['int_income'].describe()
```

yes, it is not a real depection of the incomes of the survey aprticipants because it only considers the lower brracket of the salary groups. Perhaps a better way to treat this data would be to use the median of the bracket instead of the first number of the bracket.
